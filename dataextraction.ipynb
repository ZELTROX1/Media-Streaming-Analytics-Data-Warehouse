```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("ashishkumarjayswal/movies-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd
df = pd.read_csv("movies-dataset/versions/1/movies.csv")
df.head()
```

```python
df.iloc[:,0].str.split(";",expand=True)
```

```python

df[['movie_name','rating_target_audience','genre','release_year','release_date']]=df.iloc[:,0].str.split(";",expand=True).iloc[:,0:5]
```

```python
df.head()
```

```python
df.drop('name;"rating";"genre";"year";"released";"score";"votes";"director";"writer";"star";"country";"budget";"gross";"company";"runtime"',axis=1,inplace=True)
```

```python
df.head()
```

```python
df["Unnamed: 2"].value_counts()
#this column is mostly empty, so we can drop it
```

```python
df.drop("Unnamed: 2",axis=1,inplace=True)
```

```python
#expanding the dataframe
expanded_dataframe = df.iloc[:,0].str.split(';',expand=True)
expanded_dataframe
```

```python
df[['release_year_and_country',"score","votes","director","writer","star","country","budget","gross","company","runtime"]] = expanded_dataframe
```

```python
df.head()
```

```python
#since we add extra columns, we can drop the original column
df.drop(df.columns[0],axis=1,inplace=True)
```

```python
new_col = list()
col_list = ['rating_target_audience','genre','director','writer','star','country','company','gross']
for i in col_list:
    for j in df[i]:
        if type(j)==float:
            new_col.append(j)
        else:
            new_col.append(j[1:len(j)-1])
    df[i]=new_col
    new_col=[]

```

```python
df
```

```python
new_col2 =list()
col_list2 = ['release_date','release_year_and_country']
for i in col_list2:
    if i=='release_date':
        for j in df[i]:
            if type(j)==float:
                new_col2.append(j)
            else:
                new_col2.append(j[1:])
        df[i]=new_col2
        new_col2 =[]
    else:
        for j in df[i]:
            if type(j)==float:
                new_col2.append(j)
            else:
                new_col2.append(j[:len(j)-1])
        df[i]=new_col2
        new_col2 =[]
```

```python
df.isna().sum()
```

```python
import numpy as np

df[['release_year_and_country','director','writer','star','country','company']]=df.loc[:,['release_year_and_country','director','writer','star','country','company']].replace(np.nan,'not mentioned')
```

```python
df.isna().sum()
```

```python
df[['score','votes','budget','gross','runtime']] = df.loc[:,['score','votes','budget','gross','runtime']].replace(np.nan,0)
```

```python
df['gross'] = df['gross'].replace("",0)
```

```python

df[['release_year','budget']] = df[['release_year','budget']].astype(int)
```

```python

df[['score','votes','gross']] = df[['score','votes','gross']].astype(float)
```
